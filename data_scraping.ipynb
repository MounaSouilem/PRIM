```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
url= 'https://www.mon-poeme.fr/poesies-poemes-themes/'
rep=requests.get(url)
if rep.ok:
    soup= BeautifulSoup(rep.text)
titles=soup.findAll('ul')
themes=titles[1].findAll('a')
```

```python
liens=[]
for i in themes:
    lien=i['href']
    liens.append(lien)
liens
```

```python
poemes=[]
for i in liens:
    url=i
    rep=requests.get(url)
    soup=BeautifulSoup(rep.text)
    poeme=soup.findAll('ol')
    poeme=poeme[0]
    poemes=poemes+poeme.findAll('a')

```

```python
liens_poeme=[]
for i in poemes:
    lien=i['href']
    liens_poeme.append(lien)
liens_poeme
```

```python
txt=[]
for i in liens_poeme:
    url=i
    rep=requests.get(url)
    soup=BeautifulSoup(rep.text)
    poeme=soup.findAll('div')
    txt.append(poeme[10].text)
```

```python
len(txt)
```

```python
for k in range(len(txt)):
    fichier = open("data/" + str(k+965)+".txt", "a")
    fichier.write(txt[k])
    fichier.close()
```
